```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from os.path import dirname as up
```

```python
# paths
one_up = up(os.getcwd())
save_figs = os.path.join(one_up,'data','results')
```

```python
# Load data with numpy, skipping the 'Samples' header
s1 = np.loadtxt("../data/sensor1Data.txt", skiprows=1, delimiter="\n", unpack=False)
s2 = np.loadtxt("../data/sensor2Data.txt", skiprows=1, delimiter="\n", unpack=False)
```

```python
# check out data from sensor 1
plt.plot(s1)
```

```python
# check out data from sensor 1
plt.plot(s2, c='red')
```

Miscellaneous images used in final report

```python
plt.plot(s1[5:25],color='blue', linewidth=0.7)
plt.plot(s2[50082:50102],color='red', linewidth=0.7)
plt.scatter(x=np.linspace(0,19,20),y=s1[5:25],color='blue',marker='x')
plt.scatter(x=np.linspace(0,19,20),y=s2[50082:50102],color='red',marker='o')


plt.scatter(x=np.linspace(13,14,1),y=s1[18:19],color='blue',marker='s',s=100)
plt.scatter(x=np.linspace(18,19,1),y=s2[50100:50101],color='red',marker='s',s=100)

plt.scatter(x=np.linspace(13,18,5),y=np.ones(5)*(s1[18:19]),color='black',s=10)

plt.xticks([0,10,20],fontsize=20)
plt.yticks([-0.05,0.05,0.15],fontsize=20)

plt.xlabel('Point',fontsize=20)
plt.ylabel('Signal amplitude',fontsize=20)

plt.title('Two lagged signals',fontsize=20)
plt.savefig(os.path.join(save_figs,'two_lagged_images.pdf.pdf'), bbox_inches='tight', dpi=100)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

ax1.plot(s1[0:200],linestyle='--')
ax1.plot(s2[50082:50282]+0.05,color='red')

ax1.set_yticks([-0.5, -0.25, 0])
ax1.set_xticks([0, 50, 100, 150, 200])
ax2.set_xticks([0, 50, 100, 150, 200])

ax1.set_yticklabels(['-0.5', '-0.25', '0'], fontsize=20)
ax1.set_xticklabels(['0', '50', '100', '150', '200'], fontsize = 20)
ax2.set_xticklabels(['50082', '50132', '50182', '50232', '50282'], fontsize = 20)

ax1.set_ylabel('Signal amplitude', fontsize = 20)
ax2.set_xlabel('Points (Sensor 2)', fontsize = 20)
ax1.set_xlabel('Points (Sensor 1)', fontsize = 20)

plt.title('Aligned signals\n', fontsize=20)
plt.savefig(os.path.join(save_figs,'1D_matched_signals.pdf'), bbox_inches='tight', dpi=100)
```

```python
s1_padded=np.pad(s1,(50081,0), mode='constant',constant_values=(0,0))
s2_padded=np.pad(s2,(0,50081), mode='constant',constant_values=(0,0))

plt.plot(s1_padded,color='red', alpha=1)
plt.plot(s2_padded,color='blue', alpha=0.5)



plt.xticks([0,50000,100000,150000,200000],fontsize=20)
plt.yticks([-1, -0.5, 0, 0.5, 1],fontsize=20)

ax = plt.gca()
ax.set_xticklabels(['0','50k','100k','150k','200k']) 

plt.xlabel('Point',fontsize=20)
plt.ylabel('Signal amplitude',fontsize=20)


plt.title('Signal 1 (red) overlaid with Signal 2 (blue)',fontsize=20)
plt.savefig(os.path.join(save_figs,'1D_matched_signals_all_signals.pdf'), bbox_inches='tight', dpi=100)
```

```python
plt.plot(s1[0:200])
plt.plot(s2[0:200]+0.05,color='red')
plt.plot(s1[0:200]*s2[0:200],color='black')

plt.xticks([0,50,100,150,200],fontsize=20)
plt.yticks([-0.5, -0.25, 0],fontsize=20)

plt.xlabel('Point',fontsize=20)
plt.ylabel('Signal amplitude',fontsize=20)

plt.title('Cross correlation of mismatched signals',fontsize=20)
plt.savefig(os.path.join(save_figs,'xc_1D_mismatched_signals.pdf'), bbox_inches='tight', dpi=100)
```

```python
plt.plot(s1[0:200])
plt.plot(s2[50082:50282]+0.05,color='red')
plt.plot(s1[0:200]*s2[50082:50282],color='black')

plt.xticks([0,50,100,150,200],fontsize=20)
plt.yticks([-0.5, -0.25, 0],fontsize=20)

plt.xlabel('Point',fontsize=20)
plt.ylabel('Signal amplitude',fontsize=20)

plt.title('Cross correlation of matched signals',fontsize=20)
plt.savefig(os.path.join(save_figs,'xc_1D_matched_signals.pdf'), bbox_inches='tight', dpi=100)
```
